```python
def average_round4(a_list):
    if len(a_list) > 0:
        return round(sum(a_list)/len(a_list), 4)
    else:
        return float(0)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# set your own account!!
cid = ''
secret = ''

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
import pandas as pd
from tqdm import tqdm
import time
```

```python
# create empty lists where the results are going to be stored

# 1. remember to run this initialization for the next round if you want multiple csv files for 1 year
# 2. or you can directly run the next main function and add data in these lists, 
# but remember to remove the extra data for the album information's lists!!!!

# album information
album_ID = []
album_Name = []
album_Release_Date = []
album_Artists = []
album_Total_tracks = []
album_popularity = []

# average of tracks' features
Duration_ms = [] 
Danceability = []
Energy = []
Key = []
Speechiness = []
Acousticness = []
Instrumentalness = []
Liveness = []
Valence = []
Tempo  = []
Time_Signature = []
Tracks_average_popularity = []
```

```python
# 0 is the starting point
# 100 is the ending point
# Starting from 0, but this value needs to be manipulate once any error happens.
# It should start from the (i for-loop * 50 + i trail) (the two i can be different!)
# 50 is the largest record we can get from one request each time, don't change it
for i in tqdm(range(0, 100, 50)):


    # change the year based on our task assignment
    album_results = sp.search(q = 'year: 2009', type = 'album', limit = 50, offset = i)
    
    for i, album_item in enumerate(album_results['albums']['items']):
        print("This is the %i trail in this for-loop." % i)

        time.sleep(1)
        
        total_tracks = album_item['total_tracks']
        print("Total tracks:", total_tracks)
        if total_tracks <= 24:
            album_Total_tracks.append(album_item['total_tracks'])
            
            album_ID.append(album_item['id'])
            album_Name.append(album_item['name'])
            album_Release_Date.append(album_item['release_date'])
            album_Artists.append(album_item['artists'][0]['name'])

            if sp.album(album_item['uri']) is not None:
                target = sp.album(album_item['uri'])
                album_popularity.append(target['popularity'])
            else:
                album_popularity.append('-')

            album_uri = album_item['uri']

            # initiating all the lists for storaging values of different features
            track_uri = []
            T_Duration_ms = [] 
            T_Danceability = []
            T_Energy = []
            T_Key = []
            T_Speechiness = []
            T_Acousticness = []
            T_Instrumentalness = []
            T_Liveness = []
            T_Valence = []
            T_Tempo  = []
            T_Time_Signature = []
            T_popularity = []

            for track_item in sp.album_tracks(album_uri)['items']:
                time.sleep(1)
                track_uri.append(track_item['uri'])

            for audio_item in tqdm(sp.audio_features(tracks=track_uri)):
                time.sleep(1)

                if audio_item is not None:
                
                    duration = audio_item['duration_ms']
                    if duration is not None:
                        T_Duration_ms.append(duration)
                    
                    danceability = audio_item['danceability']
                    if danceability is not None:
                        T_Danceability.append(danceability) 
                        
                    energy = audio_item['energy']
                    if energy is not None:
                        T_Energy.append(energy)
                    
                    key = audio_item['key']
                    if key is not None:
                        T_Key.append(key)
                        
                    speechiness = audio_item['speechiness']
                    if speechiness is not None:
                        T_Speechiness.append(speechiness) 
                    
                    acousticness = audio_item['acousticness']
                    if acousticness is not None:
                        T_Acousticness.append(acousticness) 
                        
                    instrumentalness = audio_item['instrumentalness']
                    if instrumentalness is not None:
                        T_Instrumentalness.append(instrumentalness) 
                 
                    liveness = audio_item['liveness']
                    if liveness is not None:
                        T_Liveness.append(liveness) 
                    
                    valence = audio_item['valence']
                    if valence is not None:
                        T_Valence.append(valence)  
                        
                    tempo = audio_item['tempo']
                    if tempo is not None:
                            T_Tempo.append(tempo) 
                    
                    time_signature = audio_item['time_signature']
                    if time_signature is not None:
                        T_Time_Signature.append(time_signature) 

            for uri in track_uri:
                track = sp.track(uri)
                time.sleep(1)
                
                if track is not None:
                    track_popularity = track['popularity']
                    if track_popularity is not None:
                        T_popularity.append(track_popularity)
        
            Duration_ms.append(average_round4(T_Duration_ms))
            Danceability.append(average_round4(T_Danceability))
            Energy.append(average_round4(T_Energy))
            Key.append(average_round4(T_Key))
            Speechiness.append(average_round4(T_Speechiness))
            Acousticness.append(average_round4(T_Acousticness))
            Instrumentalness.append(average_round4(T_Instrumentalness))
            Liveness.append(average_round4(T_Liveness))
            Valence.append(average_round4(T_Valence))
            Tempo.append(average_round4(T_Tempo))
            Time_Signature.append(average_round4(T_Time_Signature))
            Tracks_average_popularity.append(average_round4(T_popularity))

            time.sleep(3)

         
```
